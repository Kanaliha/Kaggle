```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

# Load Data

```python
data = pd.read_csv('../input/tabular-playground-series-jul-2022/data.csv',index_col='id')
sample = pd.read_csv('../input/tabular-playground-series-jul-2022/sample_submission.csv',index_col='Id')

display(data,sample)
```

# Model

```python
%%time
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 7, random_state=2)
kmeans.fit(data)
```

```python
display(kmeans.labels_)

submission = pd.DataFrame(index=data.index)
submission['Predicted'] = kmeans.labels_

submission
```

# Submission

```python
submission.to_csv('Submission.csv')
```
